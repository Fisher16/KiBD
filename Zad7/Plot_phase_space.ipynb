```python
import numpy as np

lines=open('collected.tsv').read().split('\n')[:-1]
ports,seqs=zip(*[list(map(int,line.split())) for line in lines])

ports,seqs=[np.array(lst) for lst in [ports,seqs]]
```

```python
#Based on: https://lcamtuf.coredump.cx/oldtcp/tcpseq.html
def gen_phase_space(arr):
    x=arr[2:-1]-arr[3:]
    y=arr[1:-2]-arr[2:-1]
    z=arr[:-3]-arr[1:-2]
    return x,y,z
```

```python
import pandas as pd
df=pd.DataFrame(zip(ports,seqs))
df.columns=['ports','seqs']
df.describe()
```

```python
df.hist(bins=100)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Phase space of random ports

```python
x,y,z=gen_phase_space(ports)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y,z,alpha=.2,s=5)
plt.title("Phase plot of random ports")
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection='3d')
ax.scatter(x,y,z, marker='o', s=5,alpha=.2)
ax.view_init(45,45)
```

## Phase space of random initial sequences

```python
x,y,z=gen_phase_space(seqs)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y,z,alpha=.2,s=5)
plt.title("Phase plot of random initial sequences")
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection='3d')
ax.scatter(x,y,z, marker='o', s=5,alpha=.2)
ax.view_init(45, 45)
```
